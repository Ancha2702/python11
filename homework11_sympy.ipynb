```python
from sympy import *
```

Вычисление функции,root-корни

```python
x = Symbol('x')
y = Symbol('y')
y = simplify((sin(x)/2 + cos(x)/2) ** 2 * sin(x))
func = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
root=solveset(func)
root
```

Определить промежутки, на котором f>0

```python
print(solve(func> 0,x))
```

Определить промежутки, на котором f < 0

```python
print(solve(func < 0,x))
```

Находим период функции

```python
periodicity(func, x)
```

```python
Построение график: синий- исходная функция, оранжевый - производная 
```

```python
graf = plot(func,pr, (x,-1, 2*pi))
```

Вычисляем вершину:
1. Нахождение производной

```python
pr=diff(func)
pr
```

2. Определяем корни производной:

```python
root_diff=solve(pr)
root_diff
```


Найти интервалы, на которых функция возрастает

```python
solve([x >=0, x <= periodicity(y, x), y], x)
```


Найти интервалы, на которых функция убывает

```python
solve([x <0, x <= periodicity(y, x), y], x)
```
